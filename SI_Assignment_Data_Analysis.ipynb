```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm #for calcuating the normal
```

```python
data = np.random.uniform(0,1,10)#(when a=0,b=1,n=10)it give 10 values with each distributed unifromally
```

```python
sample_mean = data.mean()
```

```python
total_sample_mean = []#(for storing the values of sample_mean )

for i in range(1000):#(calculating the unifrom distribution for 1000 times)
    data = np.random.uniform(0,1,10)
    total_sample_mean.append(data.mean())#(Addending the values of each mean in the total_sample_mean)

x = np.linspace(0,1,100)#(creating the 100 values between 0,1 with an equal increament)

a = np.array(total_sample_mean).mean()#(Converting the list into array for calucating the mean)

b = np.array(total_sample_mean).std()#(converrting the list into array for calcutaing the standard devitaion)

c=norm.pdf(x,loc=a,scale=b)

plt.hist(total_sample_mean,bins='fd',density=True)

plt.plot(x,c)
plt.show()
```

### From the above Graph we can see that the analytical form of the distribution and the histrogram shows has a good match.

```python
data = np.random.normal(1.5,2,3) # (Calculating the normal distrubtion mu=1.5,standard_devitaion=3,n=3)
```

```python
sample_mean = data.mean()#(Taking the mean of the values which we got from the above distrubtion)
```

```python
sigma = 2
n=3
std_error = sigma/np.sqrt(n)#(calculating the value of standard error)

#(Considering the z* value as 2 and it is calculated from the Z* formula for 95% confidence interval )

confidence_interval = (sample_mean - 2*std_error , sample_mean + 2*std_error)
```

```python
confidence_interval
```

### Performering the same for 10000 times 

```python
Total_confidence_interval= []
mu=1.5
sigma = 2 
std_error = sigma/np.sqrt(n)
Confidence_interval_count=0
```

```python
for i in range (10000):
    data = np.random.normal(1.5,2,3)# (where mu=1.5,standard_deviation=2,n=3)
    
    sample_mean = data.mean()#Calculating the sample from the data
    
    confidence_interval = (sample_mean - 2*std_error , sample_mean + 2*std_error)
    #(Confidence interval is calculated with sample_mean,std_error as calculated above)
    
    Total_confidence_interval.append(confidence_interval)
    #(appending the value of confidence_interval in Total_confidence_interval)
    
    if Total_confidence_interval[i][0] < mu < Total_confidence_interval[i][1]:
        #(Checking how many values of confidence_interval traps the mu)
         Confidence_interval_count+=1
            
Confidence_interval_count            
```

```python
Confidence_interval_count_percent= Confidence_interval_count*100/len(Total_confidence_interval)
print(f'The value of the confidence interval which traps the me  = {Confidence_interval_count} & the % of the Confidence_interval_count:{Confidence_interval_count_percent} %')
```
