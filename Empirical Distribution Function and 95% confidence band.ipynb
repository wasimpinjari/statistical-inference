### For the waiting time data, construct the Empirical Distribution Function and 95% confidence band, and plot these on the same plot.

```python
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
ecdf =ECDF(data)
```

### Ploting the emperical distribution function

```python
plt.plot(ecdf.x,ecdf.y,label=("ECDF"))
plt.legend(loc='best')
plt.title("Emperical Distribution")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.show()
```

```python
alpha=0.05  #level of significances 
n=len(data)
e_n=np.sqrt((1/(2*n))*(np.log(2/alpha)))
e_n
```

```python
x=np.linspace(min(data),max(data)+1,n)
```

```python
x.shape
```

```python
data.shape
```

```python
ecdf=ECDF(data)
ecdf
```

```python
a=ecdf(x)
a
```

```python
upperbond=[]
lowerbond=[]
for i in range(n):
    l=max((a[i] - e_n),0)
    u=min((a[i] + e_n),1)
    upperbond.append(u)
    lowerbond.append(l)
      
```

### ploting confidence band of EDF

```python
plt.plot(ecdf.x,ecdf.y,label='ECDF')
plt.fill_between(x, lowerbond, upperbond, alpha=0.5, label='Confidence Band')
plt.legend(loc='best')
plt.title("Confidence band")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.show()
```

### Estimate the mean waiting time, and construct the normal-based 95% condence interval for the mean.

```python
from scipy.stats import t
```

### as we don't know the standard deviation of population we use t-distribution to calculate confidence interval

```python
t=t.ppf(1-alpha/2 ,df=n-1)
```

```python
s_mean=data.mean()
```

```python
s_mean
```

```python
std_error = data.std(ddof=1)
std_error
```

```python
confidence_interval= (s_mean-t*std_error,s_mean+t*std_error)
print("The 95 % Confidence interval is:",confidence_interval)
```

### Now estimate the median, and construct the 95% condence interval for the median using 1000 boot-strap replications.

```python
from sklearn.utils import resample
```

```python
sample_median=[]
for i in range(1000):
    R=resample(data)
    M=np.median(R)
    sample_median.append(M)
```

```python
sample_median
```

```python
s_mean2=np.mean(sample_median)
std_error =np.std(sample_median ,ddof=1)
```

```python
confidence_interval= (s_mean2-t*std_error,s_mean2+t*std_error)
print("The 95 % Confidence interval is:",confidence_interval)
```
